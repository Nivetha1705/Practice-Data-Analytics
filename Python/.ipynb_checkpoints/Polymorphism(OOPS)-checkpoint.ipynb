POLYMORPHISM is a greek word meaning "One having many forms"
Every object should have the same method name only
Polymorphism can be achieved by Duck Typing - By giving same method name

```python
#here the function animal works for 3 different classes - Dog,Cat and Tiger (1 Behaves as many)
class Dog:
    def sound(self):
        print("Dog barks")
        
class Cat:        
    def sound(self):
        print("Cat meows")

class Tiger:        
    def sound(self):
        print("Tiger roars")
        
def animal(a):
    a.sound()

d=Dog()
c=Cat()
t=Tiger()
animal(d)
animal(c)
animal(t)
```

```python
#we can see here all are same method name - takeoff,fly,land. So we could easily achieve polymorphism by duck typing.
class Cargoplane:
    def takeoff(self):
        print("CargoPlane Takeoff")
    def fly(self):
        print("CargoPlane flies")
    def land(self):
        print("CargoPlane lands")
class Passengerplane:
    def takeoff(self):
        print("PassengerPlane Takeoff")
    def fly(self):
        print("PassengerPlane flies")
    def land(self):
        print("PassengerPlane lands")
class Fighterplane:
    def takeoff(self):
        print("FighterPlane Takeoff")
    def fly(self):
        print("FighterPlane flies")
    def land(self):
        print("FighterPlane lands")

def plane(a):
    a.takeoff()
    a.fly()
    a.land()

cp=Cargoplane()
pp=Passengerplane()
fp=Fighterplane()
plane(cp)
plane(pp)
plane(fp)      
```

```python
class Classmatenote:
    def writenotes(self):
        print("Classmate note helps to write neatly")
    def readnotes(self):
        print("Classmate note Helps in reading")
    def notepages(self):
        print("Classmate note has enormous number of pages")

class Apsaranote:
    def writenotes(self):
        print("Apsara note helps to write neatly")
    def readnotes(self):
        print("Apsara note Helps in reading")
    def notepages(self):
        print("Apsara note has enormous number of pages")

class Sthulnote:
    def writenotes(self):
        print("Sthul note helps to write neatly")
    def readnotes(self):
        print("Sthul note Helps in reading")
    def notepages(self):
        print("Sthul note has enormous number of pages")

def Notebook(n):
    n.writenotes()
    n.readnotes()
    n.notepages()

an=Apsaranote()
cn=Classmatenote()
sn=Sthulnote()
Notebook(an)
Notebook(cn)
Notebook(sn)       
```

```python
class Samsung:
    def start(self):
        print("Samsung Mobile starts")
    def works(self):
        print("Samsung mobile works more than 10 years")
class Apple:
    def start(self):
        print("Apple Mobile starts")
    def works(self):
        print("Apple mobile works more than 10 years")
class Vivo:
    def start(self):
        print("Vivo Mobile starts")
    def works(self):
        print("Vivo mobile works more than 10 years")
def Mobile(m):
    m.start()
    m.works()
s=Samsung()
a=Apple()
v=Vivo()
Mobile(s)
Mobile(a)
Mobile(v)

    
```

```python
class Schoolbags:
    def carrythings(self):
        print("School Bags are used to carry things")
    def carrynotes(self):
        print("School Bags are used to carry notes and books")
class Handbags:
    def carrythings(self):
        print("Hand bags are used to carry essential things")
    def carrynotes(self):
        print("Hand bags are used to carry notes and books")
class Totebags():
    def carrythings(self):
        print("Tote Bags are use to carry water bottles and things")
    def carrynotes(self):
        print("Tote bags are used to carry notes and books")
def Bags(b):
    b.carrythings()
    b.carrynotes()
sb=Schoolbags()
hb=Handbags()
tb=Totebags()
Bags(sb)
Bags(hb)
Bags(tb)
```
