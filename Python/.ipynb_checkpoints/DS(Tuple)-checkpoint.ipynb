About TUPLE:
1.Tuple is immutable --> it is not changeable
2.Tuple can be created using () paranthesis
3.Tuple is fixed or static array
4.Both Homogenous and Heterogenous values can be stored in Tuple
5.Tuple can hold duplicate values
6.Indexing is possible in Tuple
7.Slicing can also be done in tuple
8.In-built functions in tuple -->count(), index()  Other functions are not possible as it is immutable
9.common functions in tuple
10.Traversal can be done 

```python
t=(1,2,3)
print(t)
```

```python
t=(1,2,3,4)
print(t.append())   #attribute error because we cannot append as it is immutable
```

```python
#indexing in tuple
t=(1,2.56,"Tap",True)
print(t[2])
```

```python
#slicing in tuple
t=(34,24,89,573,2758,1839,27,375,9)
print(t[3:8:2])

print(t[::2])
```

TUPLE -- only consists of 2 inbuilt functions

```python
#count()
t=(4,1,2,3,4,5,4)
print(t.count(4))
```

```python
#index()
t=(46,29,246,204,284,19)
print(t.index(19))
```

Common functions that can be used in tuple as well

```python
t=(46,29,246,204,284,19)
print(max(t))
```

```python
t=(46,29,246,204,284,19)
print(min(t))
```

```python
t=(46,29,246,204,284,19)
print(type(t))
```

```python
t=(46,29,246,204,284,19)
print(sum(t))
```

```python
t=(46,29,246,204,284,19)
print(sorted(t))
```

```python
t=(46,29,246,204,284,19)
print(len(t))
```

```python
#traversal through tuple is also possible

t=(46,29,246,204,284,19)
for i in t:
    if i%2==0:
        print(i, " is even number")
    else:
        print(i, " is odd number")
```
