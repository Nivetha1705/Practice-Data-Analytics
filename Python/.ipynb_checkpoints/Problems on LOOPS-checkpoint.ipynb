PROBLEMS ON LOOPS

```python
#Print the Factors
n=int(input())
for i in range(1,n+1):
    if n%i==0:
        print(i, end=" ")
```

```python
#Count the Factors
n=int(input())
count=0
for i in range(1,n+1):
    if n%i==0:
        count+=1
print(count)    
        
```

```python
#Prime Check
import math
a=int(input())
is_prime=True
if a<=1:
    is_prime=False
else:
    for i in range(2, int(math.sqrt(a))+1):
        if a%i==0:
            is_prime=False
            break
    print("Yes" if is_prime else "No")        
```
