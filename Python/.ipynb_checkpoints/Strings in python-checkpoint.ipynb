STRINGS IN PYTHON

1. Creation of STRINGS

```python
s= "Tap's placement record is amazing" 
print(s)
```

```python
s= """Tap's placement "record" is amazing"""
print(s)
```

```python
s= '''Tap's placement "record" is amazing'''
print(s)
```

3.Slicing in strings

```python
s='PYTHON'
print(s[1:4])
```

```python
s='PYTHON'
print(s[-5:-2])
```

```python
s="PYTHON"
print(s[0:5:2])
```

4. Concatenation of strings

```python
s1= 'TAP'
s2= "ACADEMY"
s3=s1+s2
print(s3)
```

5. Strings are immutable in python

```python
s1="his name is john"
s2=" and he has a pet"
s3=s1+s2
print(s3)
```

6. Functions in strings

```python
#LOWER
s1="NIVETHA"
s2=s1.lower()
print(s2)
```

```python
#UPPER
s1="nivetha"
s2=s1.upper()
print(s2)
```

```python
#strip
s1="  nivetha  "
s2=s1.strip()
print(s2)
```

```python
#lstrip
s1="  nivetha  "
s2=s1.lstrip()
print(s2)
```

```python
#rstrip
s1="  nivetha  "
s2=s1.rstrip()
print(s2)
```

```python
#find
s1="nivetha is doing coding practices"
s2=s1.find("doing")
print(s2)
```

```python
s1="nivetha is doing coding practices"
s2=s1.find("the")
print(s2)
```

```python
#index
s1="nivetha is doing coding practices"
s2=s1.index("coding")
print(s2)
```

```python
s1="nivetha is doing coding practices"
s2=s1.index("Doing")
print(s2)
```

```python
#replace
s1="nivetha is doing coding practices"
s2=s1.replace("nivetha", "NIVETHA")
print(s2)
```

```python
#title
s1="nivetha is doing coding practices"
s2=s1.title()
print(s2)
```

```python
#Capitalize
s1="nivetha is doing coding practices"
s2=s1.capitalize()
print(s2)
```

```python
#islower() -- to check whether the string are lower

s1="nivetha is doing coding practices"
s2=s1.islower()
print(s2)
```

```python
#isupper()
#title
s1="nivetha is doing coding practices"
s2=s1.isupper()
print(s2)
```

```python
#title
s1="Nivetha is dOing codIng Practices"
s2=s1.islower()
print(s2)
```
