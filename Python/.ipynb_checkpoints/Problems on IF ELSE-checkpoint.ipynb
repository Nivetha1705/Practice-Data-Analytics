PROBLEMS ON IF ELSE TOPIC:

```python
#Even or Odd
a=int(input())
if a%2==0:
    print("Yes")
else:
    print("No")
```

```python
#Multiples of 5
a=int(input())
for i in range(1,a+1,1):
    print(i*5, end=" ")
```

```python
#Multiples of 10
n=int(input())
if n%10==0:
    print("Yes")
else:
    print("No")
```

```python
#Two digit Number
n=int(input())
if n>=10 and n<=99:
    print("Yes")
else:
    print("No")
    
```

```python
#Three Digit Number
n=int(input())
if n>=100 and n<=999:
    print("Yes")
else:
    print("No")
```

```python
#Multiplication or Addition
n=int(input())
m=int(input())
if n%2==0 and m%2==0:
    print(m*n)
else:
    print(m+n)
```

```python
#Positive and Negative Number
n=int(input())
if n>=1:
    print("Positive")
else:
    print("Negative")
```

```python
#Largest Number of two
n=int(input())
m=int(input())
if n>m:
    print(n)
else:
    print(m)
```

```python
#Largest number of three
a=int(input())
b=int(input())
c=int(input())
if a>b and a>c:
    print(a)
elif(b>c and b>a):
    print(b)
else:
    print(c)
```

```python
#Smallest Number of Two
a=int(input())
b=int(input())
if a<b:
    print(a)
else:
    print(b)
```

```python
#Smallest number of three
a=int(input())
b=int(input())
c=int(input())
if a<b and a<c:
    print(a)
elif(b<c and b<a):
    print(b)
else:
    print(c)
```
