Label should be given to every values so that we can maintain the data properly, so dictionary can be used
Label-->Key ,  Data-->Value
Dictionary is a combination of Key value pairs
It preserves the order of insertion of key value pairs
In dict, data is stored in the form of Hash table
It doesn't store duplicate keys, whereas it can store duplicate values in it

```python
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai"}
print(ds1)
```

```python
#Duplicate keys are not stored
ds1={"Name":"Tim", "Salary":100000, "Salary":1000001 ,"City":"Chennai"}
print(ds1)
```

```python
#Duplicate values are stored
ds1={"Name":"Tim", "Salary":100000, "Salary1":20000 ,"City":"Chennai"}
print(ds1)
```
