Label should be given to every values so that we can maintain the data properly, so dictionary can be used
Label-->Key ,  Data-->Value
Dictionary is a combination of Key value pairs
It preserves the order of insertion of key value pairs
In dict, data is stored in the form of Hash table
It doesn't store duplicate keys, whereas it can store duplicate values in it
Both key and value are homogenous and heterogenous in dictionary
Dictionary is mutable
Indexing and slicing is not possible in dictionary as it is not stored in an array
It is also iterable so that we can access the values

```python
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai"}
print(ds1)
```

```python
#Duplicate keys are not stored
ds1={"Name":"Tim", "Salary":100000, "Salary":1000001 ,"City":"Chennai"}
print(ds1)
```

```python
#Duplicate values are stored
ds1={"Name":"Tim", "Salary":100000, "Salary1":20000 ,"City":"Chennai"}
print(ds1)
```

```python
#add a single data in dict
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai"}
ds1.update({"Last_name":"Cook"})
print(ds1)
```

```python
#Add multiple datas in dict
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai"}
ds1.update({"Last_name":"Cook","Status":"Single","Role":"Engineer"})
print(ds1)
```

```python
#clear()
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai"}
print(ds1.clear())
```

```python
#Copy()
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai"}
ds1.copy()
print(ds1)
```

```python
#get()
#Add multiple datas in dict
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai"}
print(ds1.get("Salary"))
```

```python
#keys()
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}
print(ds1.keys())
```

```python
#values()
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}
print(ds1.values())
```

```python
#items()
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}
ds2=ds1.items()
print(ds2)
```

```python
#pop()
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}
print(ds1.pop("Status"))
```

```python
#popitem()
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}
print(ds1.popitem())
```

```python
#Setdefault()
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}
print(ds1.setdefault("Age",25))
print(ds1)
```

```python
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}
print(ds1.setdefault("Status","Married"))
print(ds1)
```

Traversal in dictionary

```python
for i in {"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}:
    print(i)
```

```python
for i in {"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}:
    print(i)
```

```python
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}
for i in ds1.keys():
    print(i)
```

```python
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}
for i in ds1.values():
    print(i)
```

```python
ds1={"Name":"Tim", "Salary":100000, "City":"Chennai","Last_name":"Cook","Status":"Single","Role":"Engineer"}
for k,v in ds1.items():
    print(k,v)
```
