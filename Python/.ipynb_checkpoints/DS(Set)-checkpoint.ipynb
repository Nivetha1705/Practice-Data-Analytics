SETS IN PYTHON
{}
Set is mutuble, because we can perform some operations on it and the set is changeable
Both list and tuple can store duplicates and it preserves the order of insertion.
But if we need a Data Structure that do not store duplicates and does not preserve the order of insertion is SET data structure 
In set data is stored in the form of Hash Table structure and not in Array
So Indexing is not possible in Set because it's not an array
Set can store both Homogenous and Heterogenous datas
No slicing is possible in set because it's not an array structure
There exists many in-built functions for set in python

```python
s={1,2,3,4,5}
print(s)
```

```python
#Order of insertion is not preserved here
s={785,204,759,274,859,285,930,465}
print(s)
```

```python
#Set does not store duplicates
s={785,204,759,274,859,285,785,930,465,785,274}
print(s)
```

```python
#Mutable
s={785,204,759,274,859,285}
s.add(345)
print(s)
```

In-Built functions for sets in python

```python
#add(one data)
s={785,204,759,274,859,285}
s.add(100)
print(s)
```

```python
#Update(multiple datas)
s={785,204,759,274,859,285}
s.update([473,8492,2947])
print(s)
```

```python
#remove()
s={785,204,759,274,859,285}
s.remove(859)
print(s)
```

```python
#pop()    ---> last data is unpredictable as it does not preserve order of insertion
s={785,204,759,274,859,285}
s.pop()
print(s)
```

```python
#Clear()
s={785,204,759,274,859,285}
s.clear()
print(s)
```

```python
#discard()
s={785,204,759,274,859,285}
s.discard(204)
print(s)
```

D/B remove() and discard()
If suppose a dont exist in the set and if we remove it, gives error
whereas if we use discard() then it don't give error

```python
s={785,204,759,274,859,285}
s.remove(100)
print(s)
```

```python
s={785,204,759,274,859,285}
s.discard(100)
print(s)
```

```python
#Copy()
s={785,204,759,274,859,285}
s.copy()
print(s)
```

UNION
To store union in a new set 
  - union
  - | --> (pipe)

```python
#union
s1={74,29,56,10,45,38}
s2={75,94,28,56,89,73}
s3=s1.union(s2)
print(s3)
```

```python
s1={74,29,56,10,45,38}
s2={75,94,28,56,89,73}
s3=s1|s2
print(s3)
```

Instead of creating a new set and storing in it, we can update it in the existing set itself
    -- |=  
    -- update

```python
s1={74,29,56,10,45,38}
s2={75,94,28,56,89,73}
s1|=s2
print(s1)
```

```python
s1={74,29,56,10,45,38}
s2={75,94,28,56,89,73}
s1.update(s2)
print(s1)
```

INTERSECTION
To store intersection in a new set
   -- intersection
   -- & (And operator)

```python
s1={74,29,56,10,45,38}
s2={75,94,28,56,89,73}
s3=s1.intersection(s2)
print(s3)
```

```python
s1={74,29,56,10,45,38}
s2={75,94,28,56,89,73}
s3=s1&s2
print(s3)
```

Instead of creating a new set and storing in it, we can update it in the existing set itself
   -- &=
   -- intersection_update

```python
s1={74,29,56,10,45,38}
s2={75,94,28,56,89,73}
s1&=s2
print(s1)
```

```python
s1={74,29,56,10,45,38}
s2={75,94,28,56,89,73}
s1.intersection_update(s2)
print(s1)
```

DIFFERENCE
To store Difference in a new set
   --> Difference
   --> - 

```python
s1={74,29,56,10,45,38}
s2={75,56,10,29,89,73}
s3=s1.difference(s2)
print(s3)
```

```python
s1={74,29,56,10,45,38}
s2={75,94,28,56,89,73}
s3=s1-s2
print(s3)
```

Instead of creating a new set and storing in it, we can update it in the existing set itself
   -- -=
   -- difference_update

```python
s1={74,29,56,10,45,38}
s2={75,56,10,29,89,73}
s1-=s2
print(s1)
```

```python
s1={74,29,56,10,45,38}
s2={75,56,10,29,89,73}
s1.difference_update(s2)
print(s1)
```
