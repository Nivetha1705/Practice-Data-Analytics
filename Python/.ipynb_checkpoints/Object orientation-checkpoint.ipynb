3 rules to remember:
   1.World is a collection of objects
       eg. Type-Computer
            i)phone
            ii)laptop
            iii)desktop ---> These all belongs to class Computer
   2.Every object belongs to a class, class is imaginary object is real
   3.Every object has something, does something
           i)Has(coded using data types)
                State/Properties
                  i)name
                  ii)cost
                  iii)mileage
           ii)does(coded using (methods/Functions))
                  Behaviour
                   i)start
                   ii)run
                   iii)stop

```python
class Car:
    name="TATA"
    cost="20.6l"
    mileage=18.90

    def start(self):
        print("Car starts")
    def run(self):
        print("Car runs")
    def stop(self):
        print("Car stops")

c=Car()
print(c.name)
print(c.cost)
print(c.mileage)
c.start()
c.run()
c.stop()
```

```python
class Dog:
    name="Mani"
    age=3
    legs=4
    colour="Black"

    def bark(self):
        print("Mani is barking")
    def walk(self):  
        print("Mani is walking")
    def eat(self):   
        print("Mani is eating")
    def sleep(self):
        print("Mani is sleeping")

a=Dog()
print(a.name)
print(a.age)
print(a.legs)
print(a.colour)
a.bark()
a.walk()
a.eat()
a.sleep()
```

```python
class Human:
    name="Nivetha"
    age=21
    gender="Female"
    email="nivetha1755@gmail.com"

    def study(self):
        print("Nivetha is studying")
    def Tv(self):
        print("She is watching TV")
    def eat(self):
        print("Nivetha is eating chicken")
    def walk(self):
        print("Nivetha is walking right now")
n=Human()
print(n.name)
print(n.age)
print(n.gender)
print(n.email)
n.study()
n.Tv()
n.eat()
n.walk()
```

```python
class Lion:
    Type="Panther Leo"
    legs=4
    weight="250 kg"
    speed="80 km/hr"

    def hunt(self):
        print("The lion hunted a deer")
    def run(self):
        print("The lion is running at 80km/hr speed")
    def sleep(self):
        print("The lion is sleeping in the den")
c=Lion()
print(c.Type)
print(c.legs)
print(c.weight)
print(c.speed)
c.hunt()
c.run()
c.sleep()
```

```python
#Here in the function purchase, we have given the name as Vijay which is overwrittern the name Ananya. 
#So the purchase function is called and then we print the name which gives Vijay as output.
class Customer:
    name="Ananya"
    ph_number=9876543210
    address="Bangalore"

    def purchase(self):
        self.name="Vijay"
    def pay(self):
        print("Ananya paid Rs.1000")
c=Customer()
c.purchase()
print(c.name)  
c.pay()
print(c.address)
```

```python
class Customer:
    name = None
    phone_num = None
    address = None

    def setname(self, x):
        self.name = x

    def set_phone_num(self, x):
        self.phone_num = x

    def set_address(self, x):
        self.address = x

c = Customer()

n = input("Enter the Name: ")
c.setname(n)
print(c.name)

m = input("Enter the Phone Number: ")
c.set_phone_num(m)
print(c.phone_num)

o = input("Enter the Address: ")
c.set_address(o)
print(c.address)
```
