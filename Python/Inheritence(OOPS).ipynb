Inheritence is defined as acquiring variables and methods from one class to another class. That is from parent class to child class

Parent class is also called as Super class or Base class

Child class is also called as Sub class or Derived class

```python
#Duplicate code in python is called as redundant code. we should always avoid duplicate code.we can do it by inheritence
class Display1:
    def disp1(self):
        print("hi")
    def disp2(self):
        print("Hello")
class Display2:
    def disp1(self):
        print("Hi")
    def disp2(self):
        print("Hello") 
```

```python
#Now what display1 has wants to be in the Display2 class also. So we should call Display1 in display2
class Display1:
    def disp1(self):
        print("hi")
    def disp2(self):
        print("Hello")
class Display2(Display1):   #Child class inherits everything from parents class
    pass                    #We cannot leave class and methods as empty, so pass

d2=Display2()
d2.disp1()
d2.disp2()
```

```python
#Not only methods, also variables can be inherited from parents class to child class
class Display1:
    def __init__(self):
        self.name="raju"
        self.age=10
    def disp1(self):
        print("hi")
    def disp2(self):
        print("Hello")
class Display2(Display1):
    pass

d2=Display2()
d2.disp1()
d2.disp2()
print(d2.name)
print(d2.age)
```

```python
class Parent:
    def __init__(self):
        self.height=5.9
        self.colour="Brown"
        self.eyecolour="Brown"
        self.haircolour="Black"

    def readbooks(self):
        print("I love reading books")
    def nodrink(self):
        print("I dont drink alcohol")
    def nosmoke(self):
        print("I dont smoke")
class child(Parent):
    pass
c=child()
print(c.height)
print(c.colour)
print(c.eyecolour)
print(c.haircolour)
c.readbooks()
c.nodrink()
c.nosmoke()
```

```python
#If we need to change any methods in the child class, we can do it in the child class
class Parent:
    def __init__(self):
        self.height=5.8
        self.colour="Brown"
        self.eyecolour="Blue"
        self.haircolour="Black"

    def readbooks(self):
        print("I love reading books")
    def nodrink(self):
        print("I dont drink alcohol")
    def nosmoke(self):
        print("I dont smoke")
        
class child(Parent):
    def nodrink(self):
        print("I love to drink alcohol")
    def readbooks(self):
        print("I hate reading books")
            
c=child()
print(c.height)
print(c.colour)
print(c.eyecolour)
print(c.haircolour)
c.readbooks()
c.nodrink()
c.nosmoke()
```

Methods coming from parents and changing in the child class is called as  OVERRIDEN METHODS
Methods coming from parents and using the methods as it is called as INHERITED METHODS

```python
#We can also create a new method in the child class alone which is not present in the parent class. It is called as SPECIALIZED METHOD
class Parent:
    def __init__(self):
        self.height=5.8
        self.colour="Brown"
        self.eyecolour="Blue"
        self.haircolour="Black"

    def readbooks(self):
        print("I love reading books")
    def nodrink(self):
        print("I dont drink alcohol")
    def nosmoke(self):
        print("I dont smoke")
        
class child(Parent):
    def nodrink(self):
        print("I love to drink alcohol")
    def readbooks(self):
        print("I hate reading books")
    def codepython(self):                 #New method created only in the child class and do not exist in parent class(Specialized method)
        print("I love python coding")
            
c=child()
print(c.height)
print(c.colour)
print(c.eyecolour)
print(c.haircolour)
c.readbooks()
c.nodrink()
c.nosmoke()
c.codepython()
```

```python
class object1:
    def __init__(self):
        self.colour="Red"
        self.size="Oval"
        self.Brand="Boult"
    def work(self):
        print("The object works good")
    def style(self):                          #Inherited method
        print("The object looks stylish")
    def quality(self):
        print("The quality of object is too good")

class object2(object1):     #Overriden methods
    def work(self):
        print("Sometimes it dont work properly")
    def quality(self):
        print("The quality is somewhat good")
o=object2()
print(o.colour)
print(o.size)
print(o.Brand)
o.work()
o.style()
o.quality()
```

```python
class object1:
    def __init__(self):
        self.colour="Red"
        self.size="Oval"
        self.Brand="Boult"
    def work(self):
        print("The object works good")
    def style(self):
        print("The object looks stylish")
    def quality(self):
        print("The quality of object is too good")

class object2(object1):
    def work(self):
        print("Sometimes it dont work properly")
    def quality(self):
        print("The quality is somewhat good")
    def charging(self):                            #Specialized method
        print("It is super fast charging")
o=object2()
print(o.colour)
print(o.size)
print(o.Brand)
o.work()
o.style()
o.quality()
o.charging()
```
