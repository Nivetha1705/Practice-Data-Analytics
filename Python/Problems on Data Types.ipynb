PROBLEMS ON DATA TYPES USING PYTHON

```python
#Add 2 integers

m=int(input())
n=int(input())
o=m+n
print(o)
```

```python
#Adding 3 integers

m=int(input())
n=int(input())
o=int(input())
p=m+n+o
print(p)
```

```python
#Product of Three

m=int(input())
n=int(input())
o=int(input())
p=m*n*o
print(p)
```

```python
#Sum Combinations

a=int(input())
b=int(input())
c=int(input())
d=a+b
e=a+c
f=c+b
print(d)
print(e)
print(f)
```

```python
#Dollar to Rupee

a=int(input())
b=82.73
c=a*b
print(f"{c:.4f}")
```

```python
#Perimeter of rectangle

l=int(input())
b=int(input())
peri_rect=2*(l+b)
print(peri_rect)
```

```python
#Perimeter of Square 

a=int(input())
peri_square = 4*a
print(peri_square)
```

```python
#area of circle

r=int(input())
pi=3.142
area_circle=pi*r*r
print(f"{area_circle:.4f}")
```

```python
#circumference of a circle

r=int(input())
pi=3.142
cir_circle=2*pi*r
print(f"{cir_circle:.4f}")
```

```python
#celsius to fahrenheit

c=int(input())
f=c*(9/5)+32
print(f"{f:.1f}")
```

```python
#Fahrenheit to celsius

f=int(input())
c=(f-32)*5/9
print(f"{c:.4f}")
```

```python
#surface area of cylinder

r=int(input())
h=int(input())
pi=3.142
area_cylinder= 2*pi*r*(r+h)
print(f"{area_cylinder:.4f}")
```
