ABSTRACTION: Hiding the implementation and showing only the essential functionality to the users is called abstraction.

In the below example, we know that the engine starts, but we do not know how it works end to end. So we dont give any implementation details and we gave pass. 
Hence if there is only method signature and no body we should use @abstractmethod. 
When @abstractmethod is there, we should give (ABC)--(Abstract Base Class) in the class to be inherited.
This ABC should be imported from the library. import abc from ABC,abstractmethod
We can achieve abstraction by ABC and @abstractmethod.


```python
from abc import ABC,abstractmethod
class Car(ABC):
    @abstractmethod
    def startengine(self):     #method signature
        pass
    @abstractmethod
    def stopengine(self):
        pass
```
