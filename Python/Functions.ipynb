```python
#multiple functions in a single cell
# 1 - no input no output
def add():
    c=10+20
    print(c)
add()

# 2 - input , output
def add(a,b):
    return a+b
print(add(5,2))    

# 3 - input, no output
def mul(a,b):
    c=a*b
    print(c)
mul(6,6)    

# 4 - no input,output
def mul():
    return 5*6
result=mul()
print(result)
```
