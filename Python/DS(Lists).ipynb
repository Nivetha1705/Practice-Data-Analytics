1.Insertion
  --append() --> to add only one value in list
  --extend() --> to add multiple values in the list
  --insert() --> to insert data at a particular index

```python
#append
a=[100,200,300]
a.append(400)
print(a)
```

```python
#extend
a=[100,200,300]
a.extend([400,500,600])
print(a)
```

```python
a=[100,200,300]
a.append([300,400,500])
print(a)
```

```python
#insert
a=[100,200,300,400]
a.insert(0,000)
print(a)
```

2. Deletion
   --pop()  --> usually removes the last data
   --pop(1) --> using index
   --remove() --> using data
   --clear()

```python
#pop
l=[100,20,47,678]
l.pop()
print(l)
```

```python
#pop(index)
l=[78,45,23,53,98]
l.pop(3)
print(l)
```

```python
#remove
l=[67,34,89,12,69]
l.remove(89)
print(l)
```

```python
#clear
l=[34,9,56,842,194]
l.clear()
print(l)
```

3. Searching
   --index()
   --count()

```python
#index
l=[100,200,300]
l2=l.index(200)
print(l2)
```

```python
l=[100,200,300]
l2=l.index(500)
print(l2)
```

```python
#count
l=["Raju", "Ramesh", "Ravi", "Ramesh", "Raki"]
print(l.count("Ramesh"))
```

```python
4. Re-ordering
   --sort()
   --reverse()
```

slicing in lists

```python
l=[45,78,34,23,89]
print(l[0:3])
```

```python
#traversal through the list

l=[1,34,67,89,32,35,56,79,53,66,14]
for i in l:
    if i%7==0:
        print(i, " is divisible by 7")
    else:
        print(i, " is not divisible by 7")
```
