OOPS- Object Oriented Programming System

OOPS-collection of class,object,encapsulation,inheritence,polymorphism,abstraction

4 pillars of OOPS:
Encapsulation
Inheritence
Polymorphism
Abstraction

Object is also called as Instance
The variables are stored in some memory inside the object which is called as instance variables

ENCAPSULATION:
The data which is carried by the variable in the object is valubale. so we have to protect it by giving security using __(double underscores)
so that it prevents the direct accessing of the variables. We can access it in a controlled way by using setters and getters methods which helps in setting the data and getting the data

```python
#Using 3 setters and 3 getters 
class Employee:
    def __init__(self):
        self.__name=None
        self.salary=0
        self.dept=None
    #setters    
    def setname(self,name):
        self.name=name
    def setsalary(self,salary):
        self.salary=salary
    def setdept(self,dept):
        self.dept=dept

    #getters
    def getname(self):
        return self.name
    def getsalary(self):
        return self.salary
    def getdept(self):    
        return self.dept
        
#object
e=Employee()
e.setname("Nivetha")
e.setsalary(70000)
e.setdept("IT")
print(e.getname())
print(e.getsalary())
print(e.getdept())      
```

```python
#Salary should not be set as negative. We can also use decision making in the setters function as it could have the ability to control the way of setting the data in the object
class Employee:
    def __init__(self):
        self.__name=None
        self.salary=0
        self.dept=None
    #setters    
    def setname(self,name):
        self.name=name
    def setsalary(self,salary):
        if salary>0:
            self.salary=salary
        else:
            salary=0
    def setdept(self,dept):
        self.dept=dept

    #getters
    def getname(self):
        return self.name
    def getsalary(self):
        return self.salary
    def getdept(self):    
        return self.dept
        
#object
e=Employee()
e.setname("Nivetha")
e.setsalary(-70000)
e.setdept("IT")
print(e.getname())
print(e.getsalary())
print(e.getdept())      
```

```python
#Even if multiple datas exist we can use only one setter instead of using 3 setters separately. In one setter itself we could pass all parameters.
class Employee:
    def __init__(self):
        self.__name=None
        self.salary=0
        self.dept=None
    #setters    
    def setdata(self,name,salary,dept):
        self.name=name
        self.salary=salary
        self.dept=dept

    #getters
    def getname(self):
        return self.name
    def getsalary(self):
        return self.salary
    def getdept(self):    
        return self.dept
        
#object
e=Employee()
e.setdata("Sharni",560000,"IT")
print(e.getname())
print(e.getsalary())
print(e.getdept())      
```
